```python
# Cell 1: Setup (works on Colab CPU or GPU)
!pip -q install -U tensorflow_datasets pillow

import os, math, numpy as np, matplotlib.pyplot as plt
import PIL.Image, PIL.ImageDraw, PIL.ImageFont
import tensorflow as tf
import tensorflow_datasets as tfds

print("TensorFlow:", tf.__version__)
print("TFDS:", tfds.__version__)
print("Devices:", tf.config.list_physical_devices())
tf.random.set_seed(42)
np.random.seed(42)
```

```python
# Cell 2: Globals & drawing helpers

IM_WIDTH = 75
IM_HEIGHT = 75

def draw_bounding_boxes_on_image_array(image, boxes, colors=None, thickness=2):
    """
    image: (H,W) or (H,W,1) uint8 array
    boxes: (N, 4) [xmin, ymin, xmax, ymax] in normalized coords
    colors: list like ["green","red"]
    """
    if image.ndim == 3:
        image = image.squeeze(-1)
    img = PIL.Image.fromarray(image.astype(np.uint8), mode="L").convert("RGBA")
    draw = PIL.ImageDraw.Draw(img)
    im_w, im_h = img.size

    if boxes is None:
        boxes = np.zeros((0,4), dtype=np.float32)
    for i, box in enumerate(boxes):
        xmin, ymin, xmax, ymax = [float(v) for v in box]
        left, top   = xmin * im_w, ymin * im_h
        right, bot  = xmax * im_w, ymax * im_h
        col = (colors[i] if (colors and i < len(colors)) else "red")
        draw.rectangle([left, top, right, bot], outline=col, width=thickness)
    return np.array(img)
```

```python
# Cell 3: Display & metric helpers

def intersection_over_union(pred_box, true_box):
    """
    pred_box, true_box: arrays (N,4) [xmin,ymin,xmax,ymax]
    returns IoU shape (N,1)
    """
    xmin_p, ymin_p, xmax_p, ymax_p = np.split(pred_box, 4, axis=1)
    xmin_t, ymin_t, xmax_t, ymax_t = np.split(true_box, 4, axis=1)

    smooth = 1e-10
    xmin_o = np.maximum(xmin_p, xmin_t)
    ymin_o = np.maximum(ymin_p, ymin_t)
    xmax_o = np.minimum(xmax_p, xmax_t)
    ymax_o = np.minimum(ymax_p, ymax_t)

    inter = np.maximum(xmax_o - xmin_o, 0) * np.maximum(ymax_o - ymin_o, 0)
    area_p = (xmax_p - xmin_p) * (ymax_p - ymin_p)
    area_t = (xmax_t - xmin_t) * (ymax_t - ymin_t)
    union = area_p + area_t - inter

    iou = (inter + smooth) / (union + smooth)
    return iou

def display_digits_with_boxes(digits, predictions, labels,
                              pred_bboxes=None, bboxes=None, iou=None,
                              title="", iou_threshold=0.6, n=5):
    """
    digits: (N,75,75,1) float [0..1]
    predictions, labels: (N,)
    pred_bboxes, bboxes: (N,4) normalized
    """
    N = len(labels)
    n = min(n, N)
    idx = np.random.choice(N, size=n, replace=False)

    fig = plt.figure(figsize=(n*3.5, 3.5))
    plt.suptitle(title)
    for j, i in enumerate(idx):
        ax = fig.add_subplot(1, n, j+1)

        img_u8 = (digits[i].squeeze() * 255.0).astype(np.uint8)
        boxes = []
        colors = []
        # Draw GT first (green), then prediction (red)
        if bboxes is not None:
            boxes.append(bboxes[i])
            colors.append("green")
        if pred_bboxes is not None:
            boxes.append(pred_bboxes[i])
            colors.append("red")

        drawn = draw_bounding_boxes_on_image_array(img_u8, np.array(boxes) if boxes else None, colors=colors)
        ax.imshow(drawn)
        ax.set_xticks([]); ax.set_yticks([])
        ax.set_xlabel(f"pred {predictions[i]} / true {labels[i]}")
        if iou is not None:
            color = "black" if iou[i][0] >= iou_threshold else "red"
            ax.text(0.02, -0.25, f"IoU: {iou[i][0]:.2f}", transform=ax.transAxes, color=color)
    plt.show()

def plot_history(history, metric, title):
    plt.figure(figsize=(6,4))
    plt.title(title)
    plt.plot(history.history[metric], label=metric)
    val_key = f"val_{metric}"
    if val_key in history.history:
        plt.plot(history.history[val_key], label=val_key)
    plt.legend()
    plt.xlabel("epoch"); plt.ylabel(metric)
    plt.show()
```

```python
# Cell 4: Datasets

def read_image_tfds(image, label):
    # Random top-left for a 28x28 digit inside a 75x75 canvas
    xmin = tf.random.uniform((), 0, IM_WIDTH - 28, dtype=tf.int32)
    ymin = tf.random.uniform((), 0, IM_HEIGHT - 28, dtype=tf.int32)

    image = tf.reshape(image, (28, 28, 1))
    image = tf.image.pad_to_bounding_box(image, ymin, xmin, IM_HEIGHT, IM_WIDTH)
    image = tf.cast(image, tf.float32) / 255.0

    xmin_f = tf.cast(xmin, tf.float32) / IM_WIDTH
    ymin_f = tf.cast(ymin, tf.float32) / IM_HEIGHT
    xmax_f = (tf.cast(xmin, tf.float32) + 28.0) / IM_WIDTH
    ymax_f = (tf.cast(ymin, tf.float32) + 28.0) / IM_HEIGHT

    bbox = tf.stack([xmin_f, ymin_f, xmax_f, ymax_f], axis=0)
    label_oh = tf.one_hot(label, 10)
    return image, (label_oh, bbox)

strategy = tf.distribute.get_strategy()
BATCH_SIZE = 128 * strategy.num_replicas_in_sync

# Train: MNIST train split
train_ds = tfds.load("mnist", split="train", as_supervised=True)
train_ds = train_ds.map(read_image_tfds, num_parallel_calls=tf.data.AUTOTUNE)
train_ds = train_ds.shuffle(5000, reshuffle_each_iteration=True).repeat()
train_ds = train_ds.batch(BATCH_SIZE, drop_remainder=True).prefetch(tf.data.AUTOTUNE)

# Validation: MNIST test split
val_ds = tfds.load("mnist", split="test", as_supervised=True)
val_ds = val_ds.map(read_image_tfds, num_parallel_calls=tf.data.AUTOTUNE)
val_ds = val_ds.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

# Grab a NumPy batch for quick visualization later
sample_digits, (sample_labels_oh, sample_bboxes) = next(iter(val_ds))
val_digits_np = sample_digits.numpy()
val_labels_np = np.argmax(sample_labels_oh.numpy(), axis=1)
val_bboxes_np = sample_bboxes.numpy()

display_digits_with_boxes(val_digits_np, val_labels_np, val_labels_np,
                          pred_bboxes=None, bboxes=val_bboxes_np, iou=None,
                          title="Validation samples (GT only)", n=5)
```

```python
# Cell 5: Model

def feature_extractor(inputs):
    x = tf.keras.layers.Conv2D(16, 3, activation='relu')(inputs)
    x = tf.keras.layers.AveragePooling2D(2)(x)
    x = tf.keras.layers.Conv2D(32, 3, activation='relu')(x)
    x = tf.keras.layers.AveragePooling2D(2)(x)
    x = tf.keras.layers.Conv2D(64, 3, activation='relu')(x)
    x = tf.keras.layers.AveragePooling2D(2)(x)
    return x

def dense_layers(inputs):
    x = tf.keras.layers.Flatten()(inputs)
    x = tf.keras.layers.Dense(128, activation='relu')(x)
    return x

def classifier_head(inputs):
    return tf.keras.layers.Dense(10, activation="softmax", name="classification")(inputs)

def bbox_head(inputs):
    return tf.keras.layers.Dense(4, name="bounding_box")(inputs)

def build_model():
    inp = tf.keras.layers.Input(shape=(IM_HEIGHT, IM_WIDTH, 1))
    feat = feature_extractor(inp)
    dense = dense_layers(feat)
    cls_out = classifier_head(dense)
    box_out = bbox_head(dense)
    model = tf.keras.Model(inputs=inp, outputs=[cls_out, box_out])
    model.compile(
        optimizer='adam',
        loss={'classification': 'categorical_crossentropy', 'bounding_box': 'mse'},
        metrics={'classification': 'accuracy', 'bounding_box': 'mse'}
    )
    return model

with strategy.scope():
    model = build_model()

model.summary()
```

```python
# Cell 6: Train & evaluate

EPOCHS = 5  # increase if you want
train_examples = 60000
steps_per_epoch = math.ceil(train_examples / BATCH_SIZE)

history = model.fit(
    train_ds,
    epochs=EPOCHS,
    steps_per_epoch=steps_per_epoch,
    validation_data=val_ds
)

eval_dict = model.evaluate(val_ds, return_dict=True)
print("\n--- Evaluation ---")
for k, v in eval_dict.items():
    print(f"{k}: {v:.4f}")
```

```python
# Cell 7: Plots

plot_history(history, "classification_accuracy", "Classification Accuracy")
plot_history(history, "classification_loss", "Classification Loss")
plot_history(history, "bounding_box_mse", "Bounding Box MSE")
```

```python
# Cell 8: Predictions & visualization

# Use the cached NumPy batch we grabbed earlier
pred_cls_probs, pred_bboxes = model.predict(val_digits_np, batch_size=64, verbose=0)
pred_labels = np.argmax(pred_cls_probs, axis=1)

iou_vals = intersection_over_union(pred_bboxes, val_bboxes_np)
display_digits_with_boxes(val_digits_np, pred_labels, val_labels_np,
                          pred_bboxes=pred_bboxes, bboxes=val_bboxes_np,
                          iou=iou_vals, title="GT (green) vs Pred (red)", n=5, iou_threshold=0.6)
```
